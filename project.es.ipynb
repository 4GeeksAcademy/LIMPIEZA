#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO

import matplotlib.pyplot as plt

# Normalizar la variable de precios para cada población
cinturon_sur['precio_normalizado'] = cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())

# Crear una gráfica de histograma para cada población
plt.figure(figsize=(12, 8))
for poblacion, datos in cinturon_sur.groupby('level5'):
    plt.hist(datos['precio_normalizado'], bins=20, alpha=0.6, label=poblacion)

plt.title('Histograma de Precios Normalizados por Población en el Cinturón Sur de Madrid')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(title='Población', loc='upper right')
plt.grid(True)
plt.show()
```

la principal conclusion al ver el sesgo hacia la derecha, lo que puedo decir es que hay mas casas con precios normalizados más altos.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆)

```python
# TODO

import matplotlib.pyplot as plt

# Crear una figura y ejes de subgráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Lista de poblaciones a incluir en el gráfico
poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

# Iterar sobre las poblaciones y crear un gráfico de dispersión para cada una
for i, poblacion in enumerate(poblaciones):
    # Filtrar el DataFrame para la población actual
    poblacion_df = cinturon_sur[cinturon_sur['level5'] == poblacion]
    # Crear el gráfico de dispersión en el subgráfico correspondiente
    axs[i//2, i%2].scatter(poblacion_df['surface'], poblacion_df['price'], alpha=0.5)
    axs[i//2, i%2].set_title(f'Dispersión en {poblacion}')
    axs[i//2, i%2].set_xlabel('Metros cuadrados construidos')
    axs[i//2, i%2].set_ylabel('Precio')
    axs[i//2, i%2].grid(True)

# Ajustar el espacio entre los subgráficos
plt.tight_layout()
plt.show()
```
